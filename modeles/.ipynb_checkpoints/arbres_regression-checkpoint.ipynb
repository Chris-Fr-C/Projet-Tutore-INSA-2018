# Arbre de regression

Les arbres de régression sont un type de modèle particulier pour lequel nous avons un modèle prédictif discontinu.

Exemple de fonctionnement de l'algorithme :

<img src="img/exemple_arbre.png">




0. Récupération des données d'entrainement
0. Entrainement :
    1. Normalisation éventuelle
    1. Séparation du set d'entrainement et de test
    1. Creation du modèle
    1. Entrainement du modèle
0. Prediction :
    1. Normalisation éventuelle
    1. Prédiction avec les paramètres d'entrée
    1. Dénormalisation éventuelle

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sklearn
try:
    import seaborn as sns
    sns.set_style('dark')
except ImportError:
    print("On utilise seulement matplotlib")
```

```python
# =================================================
# PARAMETRES DU MODELE
XMIN = -1.
XMAX = 1.
NB_PTS = 20
RANDOM_WEIGHT = 0.1
DEGREE = 2
SIZE = (15,10) # plot size
# =================================================

x = np.linspace(XMIN,XMAX,NB_PTS).reshape(-1,1)
# On crée les variables dépendantes
# Elles vont correspondre à un polynome de second ordre avec du bruit
y = (x+0.5)*(x-0.15) + RANDOM_WEIGHT*np.random.random(x.shape)
```

```python
# Affichage des données
fig = plt.figure(figsize=SIZE)

plt.scatter(x,y,label="Points initiaux")
plt.legend()
plt.show()
```

## Creation du modèle

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x,y)
```

## Affichage du résultat

```python
x_grid = np.linspace(XMIN,XMAX,NB_PTS*10).reshape(-1,1)
y_pred = model.predict(x_grid)

fig = plt.figure(figsize=SIZE)

plt.plot(x_grid,y_pred,label="Points prédits",color="r")
plt.scatter(x,y,label="Points initiaux")
plt.legend()
plt.show()
```
