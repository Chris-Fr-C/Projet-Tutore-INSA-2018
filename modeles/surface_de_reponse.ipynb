# **Modèles de substitution**

Soit $X$ le vecteur d'entrée. Soit $Y$ le vecteur de sortie
Nous cherchons un modèle $F$ tel que $F(X)$ prédise au mieux la sortie $Y$ correspondante.
Le critère de coût peut être être varié, mais nous utiliserons ici la fonction de coût suivante, appelée Moindres carrés.
Elle est évaluée telle que :
$Cout = \sum_{i=0}^{n}{(Y_{true} - Y_{predicted})^{2}}$



# Surface de réponse

Les surfaces de réponses sont un type de modèle particulier pour lequel nous avons un modèle prédictif de la forme suivante:
$F(X)= \sum_{i=0}^{degree}({a_i\prod_{j=0}^{degree}{X_j})}$

0. Récupération des données d'entrainement
0. Entrainement :
    1. Normalisation éventuelle
    1. Séparation du set d'entrainement et de test
    1. Creation du modèle
    1. Entrainement du modèle
0. Prediction :
    1. Normalisation éventuelle
    1. Prédiction avec les paramètres d'entrée
    1. Dénormalisation éventuelle

## Generation du set de données

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sklearn
try:
    import seaborn as sns
    sns.set_style("dark")
except ImportError:
    print("On utilise seulement matplotlib")
```

```python
# =================================================
# PARAMETRES DU MODELE
XMIN = -1.
XMAX = 1.
NB_PTS = 500
RANDOM_WEIGHT = 0.2
DEGREE = 2
SIZE = (15,10) # plot size
# =================================================

x = np.linspace(XMIN,XMAX,NB_PTS).reshape(-1,1)
# On crée les variables dépendantes
# Elles von correspondre à un polynome de second ordre avec du bruit
y = (x-0.3) * (x-0.25)
# On ajoute le bruit
y = y + RANDOM_WEIGHT * np.random.random(x.shape)
```

```python
# Affichage des données
fig = plt.figure(figsize=SIZE)

plt.title("Point initiaux bruités")
plt.scatter(x,y,label="Points initiaux")
plt.legend()
plt.show()
```

## Creation du modèle

```python
# Il faut tout d'abord créer un modèle permettant de convertir le vecteur d'entrée, en vecteur polynomial
from sklearn.preprocessing import PolynomialFeatures

poly_feat = PolynomialFeatures(degree=DEGREE, include_bias=False)
x_poly = poly_feat.fit_transform(x)
```

```python
# Construction du modèle
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_poly,y)
```

## Affichage du résultat

```python
x_grid = np.linspace(XMIN,XMAX,NB_PTS*10).reshape(-1,1)
x_grid_poly = poly_feat.transform(x_grid)
y_pred = model.predict(x_grid_poly)

fig = plt.figure(figsize=SIZE)

plt.title("Prediction")
plt.plot(x_grid,y_pred,label="Points prédits",color="r")
plt.scatter(x,y,label="Points initiaux")
plt.legend()
plt.show()
```
